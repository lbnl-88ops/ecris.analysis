```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(0.001, 10, 1000)
#noise = [np.random.normal(10, 2) for _ in x]
noise = np.zeros_like(x)
m = 16
q = 8
o_center = 5
peak_x = [m/q for q in range(1, q) if m/q <= 10]
peak_y = np.zeros_like(x)
for peak in peak_x:
    height = 500*norm.pdf(x, loc=o_center, scale=2)
    #peak_y += height*norm.pdf(x, loc=peak, scale=0.1)
    peak_y += 100*norm.pdf(x, loc=peak, scale=0.1)
signal = peak_y + noise

search = np.zeros_like(x)
for peak in peak_x:
    search += 100*norm.pdf(x, loc=peak, scale=0.1)

plt.plot(x, signal, '.')
plt.plot(x, search, '.')
plt.grid()
plt.show()
```

```python
convolution = np.convolve(signal, search)
plt.plot(convolution)
plt.show()
np.argmax(convolution)
```
