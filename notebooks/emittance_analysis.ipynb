## Opening Emittance Scan Files

Emittance scan files are stored as hdf5 files with embedded data and metadata.

```python
from pathlib import Path
emittance_scan_path = Path('../../emittance_viewer/tmp/emittance_scan_1774303663.h5')
print(f'{emittance_scan_path.exists()=}')
```

```python
from ops.ecris.analysis.io.read_emittance_scan_file import load_emittance_scan
import numpy as np
es = load_emittance_scan(emittance_scan_path)
print(np.min(es.data))
print(np.max(es.data))
```

The loaded emittance scan object `EmittanceScan` contains all the parsed data and metadata

```python
# Timestamp
print(es.timestamp)
```

```python
# Scan parameters
print(f"Scan axis: {es.scan_parameters.axis}\n\n")

# All of them:
for l, v in es.scan_parameters.__dict__.items():
    print(l, ":", v)
```
